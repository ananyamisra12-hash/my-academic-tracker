```python
import random
import time

# ==========================================
# 1. INITIAL FLASHCARD DATABASE
# ==========================================
flashcards = [
    {"id": 1, "subject": "Chemistry", "question": "What is the functional group of an alcohol?", "answer": "-OH (Hydroxyl group)"},
    {"id": 2, "subject": "Physics", "question": "What is Lenz's Law used to determine?", "answer": "The direction of an induced electromotive force (EMF) or current."},
    {"id": 3, "subject": "Data Science", "question": "What does 'Overfitting' mean in Machine Learning?", "answer": "When a model learns training data noise too well, failing to generalize."}
]

# ==========================================
# 2. HELPER FUNCTIONS
# ==========================================
def print_header(title):
    print("\n" + "="*50)
    print(f" {title} ".center(50, "•"))
    print("="*50)

def show_stats():
    total = len(flashcards)
    chem = sum(1 for c in flashcards if c['subject'] == 'Chemistry')
    phys = sum(1 for c in flashcards if c['subject'] == 'Physics')
    ds = sum(1 for c in flashcards if c['subject'] == 'Data Science')
    
    print(f"📊 DECK STATUS: {total} total cards remaining")
    print(f"🧪 Chemistry: {chem} | ⚡ Physics: {phys} | 💻 Data Science: {ds}")

# ==========================================
# 3. CORE INTERACTIVE LOOP
# ==========================================
def start_flashcard_app():
    global flashcards
    
    while True:
        print_header("MAIN MENU")
        show_stats()
        print("\n[1] Review Flashcards (Study Mode)")
        print("[2] Add a Custom Flashcard")
        print("[3] Exit Application")
        
        choice = input("\nSelect an option (1-3): ").strip()
        
        if choice == "1":
            if not flashcards:
                print("\n🎉 Awesome! Your flashcard deck is completely empty. You've mastered everything!")
                continue
                
            # Take the first card in the queue
            card = flashcards.pop(0)
            
            print_header(f"STUDYING: {card['subject'].upper()}")
            print(f"❓ QUESTION:\n--> {card['question']}")
            
            input("\nPress ENTER to reveal the answer... ")
            print(f"\n💡 ANSWER:\n--> {card['answer']}")
            
            print("\nDid you remember this correctly?")
            print("[1] Hard (See it again very soon)")
            print("[2] Easy (Move it to the back of the deck)")
            
            feedback = input("Your choice (1-2): ").strip()
            
            if feedback == "1":
                # Spaced Repetition: Put it right back at the front (index 0 or 1)
                flashcards.insert(0, card)
                print("\n🔁 Card marked as Hard. It's queued up for your next turn!")
            else:
                # Easy: Send it to the back of the line
                flashcards.append(card)
                print("\n✅ Card marked as Easy. Sent to the back of the deck!")
                
            time.sleep(1)
            
        elif choice == "2":
            print_header("ADD NEW CARD")
            print("Choose Subject:")
            print("[1] Chemistry  [2] Physics  [3] Data Science")
            sub_choice = input("Select subject (1-3): ").strip()
            
            subjects = {"1": "Chemistry", "2": "Physics", "3": "Data Science"}
            chosen_subject = subjects.get(sub_choice, "General")
            
            q = input("Enter Question: ").strip()
            a = input("Enter Answer: ").strip()
            
            if q and a:
                flashcards.append({
                    "id": len(flashcards) + 1,
                    "subject": chosen_subject,
                    "question": q,
                    "answer": a
                })
                print(f"\n✨ Success! Added your new {chosen_subject} card to the loop.")
            else:
                print("\n⚠️ Error: Both fields are required. Card not added.")
                
            time.sleep(1)
            
        elif choice == "3":
            print("\n👋 Happy studying! Closing the revision panel dashboard. See ya!")
            break
        else:
            print("\n⚠️ Invalid option. Please enter 1, 2, or 3.")
            time.sleep(1)

# Run the app engine
start_flashcard_app()
```
